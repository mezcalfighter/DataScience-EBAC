```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv", index_col=1)
```

```python
fig, ax = plt.subplots()
ax.hist(df["Age"])
```

```python
fig, ax = plt.subplots()
ax.set_xticklabels(["Edades"])
ax.boxplot(df["Age"])
plt.show
```

```python
fig, ax = plt.subplots()
ax.bar(df.index.astype(int),df["Age"].astype(int))
ax.set_xlabel("Clase")
ax.set_ylabel("Edad")
plt.show
```
