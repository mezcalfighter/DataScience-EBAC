```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv("co2-mm-mlo_csv.csv")
df
```

```python
fechas = df[df["Date"] >= "2015-01-01"]

fig, ax = plt.subplots()
ax.plot(fechas["Date"],fechas["Interpolated"])
plt.show
```

```python
fechas = df[df["Date"] >= "2015-01-01"]

fig, ax = plt.subplots()
ax.plot(fechas["Date"],fechas["Interpolated"])
ax.axes.xaxis.set_ticks([]) # Para quitar las leyendas asociadas a los puntos
plt.xlabel("Datos mensuales: 2015-01-01 - 2018-09-01")
plt.ylabel("Emisiones de C02 promedio")
plt.show
```

```python
# Creacion de una funciona que grafica series en el tiempo
def grafica_series_tiempo(ejes,x,y,color,etiqueta_x,etiqueta_y):
    ejes.plot(x,y,color=color)
    ejes.set_xlabel(etiqueta_x)
    ejes.set_ylabel(etiqueta_y, color=color)
    ejes.tick_params("y", colors=color)
```

```python
fig,ax = plt.subplots()
grafica_series_tiempo(ax,fechas["Date"],fechas["Interpolated"],"blue",
                      "Datos mensuales: 2015-01-01 - 2018-09-01", "C02")
ax.axes.xaxis.set_ticks([])
plt.show
```

```python
# Incorporar otro elemento grafico en la misma figura
fig,ax = plt.subplots()
grafica_series_tiempo(ax,fechas["Date"],fechas["Interpolated"],"blue",
                      "Datos mensuales: 2015-01-01 - 2018-09-01", "C02 Interpolated")

ax2 = ax.twinx()

grafica_series_tiempo(ax2,fechas["Date"],fechas["Trend"],"red",
                      "", "C02 Trend")

ax.axes.xaxis.set_ticks([])
plt.show
```
