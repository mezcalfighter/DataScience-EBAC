```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("olympic-medals.csv")
df
```

```python
df = pd.read_csv("olympic-medals.csv", index_col=0)
df
```

```python
medallas = df.head(10)
medallas
```

```python
# Medallas de barra
fig, ax = plt.subplots()
ax.bar(medallas.index,medallas["Gold"])
plt.show
```

```python
#Corregir que los nombres no se entrelacen 
# Primer solucion, rotandolos 90 grados
fig, ax = plt.subplots()
ax.bar(medallas.index,medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de Medallas")
plt.show
```

```python
# Visualizacion de Medallas de Oro y Plata
fix, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show
```

```python
# Visualizacion de Medallas de Oro, Plata y Bronce
fix, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom = medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show
```

```python
# Visualizacion de Medallas de Oro, Plata y Bronce
fix, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom = medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
ax.legend(["Gold","Silver","Bronze"])
plt.show
```

```python
fig, ax = plt.subplots()
ax.hist(df["Gold"])
plt.show
```

```python
# Creacion de dos histogramas en un solo grafico

fix, ax = plt.subplots()
ax.hist(df["Gold"])
ax.hist(df["Silver"])
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
plt.show
```

```python
# Creacion de dos histogramas en un solo grafico

fix, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro")
ax.hist(df["Silver"], label = "Plata")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show
```

```python
# Creacion de histogramas con especificos de intervalos

fix, ax = plt.subplots()
ax.hist(df["Gold"], label="Oro", bins=[0,5,10,15,20,25,30,35,40])
ax.hist(df["Silver"], label="Plata", bins=[0,5,10,15,20,25,30,35,40])
ax.set_xlabel("# de Medallas")
ax.set_ylabel("# de Observaciones")
ax.legend()
plt.show
```

```python
# Creacion de histogramas transparentes
fix, ax = plt.subplots()
ax.hist(df["Gold"], label="Oro", bins=[0,5,10,15,20,25,30,35,40], histtype="step")
ax.hist(df["Silver"], label="Plata", bins=[0,5,10,15,20,25,30,35,40], histtype="step")
ax.set_xlabel("# de Medallas")
ax.set_ylabel("# de Obsercaciones")
ax.legend()
plt.show
```

```python
# Creacion de diagramas de caja
fig, ax = plt.subplots()
ax.boxplot([df["Gold"], df["Silver"]])
ax.set_xticklabels(["Oro","Plata"])
ax.set_ylabel("# de Medallas")
plt.show
```

```python
# Creacion de graficos de dispersion
fig, ax = plt.subplots()
ax.scatter(df["Gold"],df["Silver"])
ax.set_xticklabels(["Oro","Plata"])
ax.set_ylabel("# de Medallas")
plt.show
```

```python
# Creacion de graficos de dispersion multiples
fig, ax = plt.subplots()
ax.scatter(df["Gold"],df["Silver"], color = "red", label = "Plata")
ax.scatter(df["Gold"],df["Bronze"], color = "blue", label = "Bronce")
ax.set_xlabel("# de Medallas de Oro")
ax.set_ylabel("# de Medallas")
ax.legend()
plt.show
```

```python
# Creacion de graficos de dispersion con degradado en color
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"], c = df.index)
ax.set_xlabel("# de Medallas de Oro")
ax.set_ylabel("# de Medallas")
plt.show
```
